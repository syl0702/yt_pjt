```python
# 각 카테고리별로 translations를 한꺼번에 모으려고 할 때 사용하면 되는 코드입니다.
# 정의는 마음대로 수정하시면 됩니다!
# 이후의 코드에서 example = 부분에 combined 를 적용하지면 됩니다.
howto_translations = df[df['category_title'] == 'Howto & Style']['translations']
combined_howto_t = ' '.join(howto_translations.fillna(''))
combined_howto_t
```

```python
from konlpy.tag import Okt, Kkma
import re

# Okt와 Kkma 객체 생성
okt = Okt()
kkma = Kkma()

# 한국어 텍스트를 명사와 어근으로 한정하여 추출
def preprocess_korean_text(text):
    # Kkma를 사용하여 형태소 분석
    tokens_kkma = kkma.pos(text)

    # 명사와 어근만 추출
    selected_tokens = [word for word, pos in tokens_kkma if pos.startswith('N') or pos == 'XR']

    # 불용어를 제외한 토큰 추출
    stop_words # 불용어 목록 추가
    selected_tokens = [token for token in selected_tokens if token not in stop_words]

    return selected_tokens

# 예제 한국어 텍스트
# 카테고리별로 합친 것을 여기에 재정의해주세요!
example = df.loc[1, 'translations']

# 전처리 수행
preprocessed_tokens = preprocess_korean_text(example)

# 전처리된 결과 출력
preprocessed_tokens
```

```python
from gensim.models.ldamodel import LdaModel
from gensim.models.callbacks import CoherenceMetric
from gensim import corpora
from gensim.models.callbacks import PerplexityMetric

import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

dictionary = corpora.Dictionary([preprocessed_tokens])
```

```python
corpus = [dictionary.doc2bow(text) for text in [preprocessed_tokens]]
```

```python
num_topics = 5
chunksize = 2000
passes = 20
iterations = 400
eval_every = None

temp = dictionary[0]
id2word = dictionary.id2token

model = LdaModel(
    corpus=corpus,
    id2word=id2word,
    chunksize=chunksize,
    alpha='auto',
    eta='auto',
    iterations=iterations,
    num_topics=num_topics,
    passes=passes,
    eval_every=eval_every
)
```

```python
top_topics = model.top_topics(corpus)

avg_topic_coherence = sum([t[1] for t in top_topics]) / num_topics
print('Average topic coherence: %.4f.' % avg_topic_coherence)

from pprint import pprint
pprint(top_topics)
```
