```python
def categories(num):
    try:
        if int(num) == 1:
            return 'Film & Animation'

        elif int(num) == 2:
            return 'Autos & Vehicles'

        elif int(num) == 10:
            return 'Music'

        elif int(num) == 15:
            return 'Pets & Aminals'

        elif int(num) == 17:
            return 'Sports'

        elif int(num) == 18:
            return 'Short Movies'

        elif int(num) == 19:
            return 'Travel & Events'

        elif int(num) == 20:
            return 'Gaming'

        elif int(num) == 21:
            return 'Videoblogging'

        elif int(num) == 22:
            return 'People & Blogs'

        elif int(num) == 23:
            return 'Comedy'

        elif int(num) == 24:
            return 'Entertainment'

        elif int(num) == 25:
            return 'News & Politics'

        elif int(num) == 26:
            return 'Howto & Style'

        elif int(num) == 27:
            return 'Education'

        elif int(num) == 28:
            return 'Science & Technology'
        
    except:
        return num
```

```python
for i in range(len(df1)):
    print('index: ', i, '|', 'title: ', df1.loc[i, 'title'])
    print('category: ', df1.loc[i, 'categoryTitle'])
    print('')
    print(df1.loc[i, 'translations_ko'])
    
    cat = input("카테고리를 변경하시겠습니까?: ")
            
    if cat == 'n':
        pass
    else:
        cate = input('category: ')
        df.loc[i, 'categoryTitle'] = categories(cate)

    
    if df.loc[i, 'translations_ko'] == 'nan' or isinstance(df1.loc[i, 'translations_ko'], float):
        clear_output(wait=True)
        continue
        
    else:
        modifying = input('스크립트를 변경하시겠습니까?:')
    
        if modifying == 'n':
            clear_output(wait=True)
            continue
            
        else:
            scripts = input("변경될 스크립트:")

            if scripts:
                df1.loc[i, 'translations_ko'] = scripts
                clear_output(wait=True)

            else:
                script = input('스크립트를 입력해주세요: ')
                df1.loc[i, 'translations_ko'] = script
                
    clear_output(wait=True)
```
