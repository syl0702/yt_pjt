```python
from googleapiclient.discovery import build

# API 키를 설정합니다.
api_key = "AIzaSyA701JVS8TN7tjKo9H8EjzdYYkPT9cdxx8"

# YouTube API 클라이언트를 빌드합니다.
youtube = build("youtube", "v3", developerKey=api_key)

# 비디오 ID를 지정합니다.
video_id = "io2_CXLBb7c"

# 비디오의 자막을 가져옵니다.
captions = youtube.captions().list(
    part="snippet",
    videoId=video_id
).execute()

# 원하는 언어의 자막을 추출합니다.
desired_language = "ko"  # 한국어 자막
script_text = ""

for caption in captions.get("items", []):
    if caption["snippet"]["language"] == desired_language:
        caption_id = caption["id"]
        caption_data = youtube.captions().download(
            id=caption_id,
            tfmt="srt"
        ).execute()
        script_text = caption_data

# script_text에 추출된 자막이 저장됩니다.
print(script_text)
```

```python
import requests

# YouTube 비디오 ID와 API 키를 설정합니다.
video_id = "io2_CXLBb7c"
api_key = "AIzaSyA701JVS8TN7tjKo9H8EjzdYYkPT9cdxx8"

# 자막 트랙 정보를 얻기 위한 API 엔드포인트를 설정합니다.
track_info_endpoint = f"https://www.googleapis.com/youtube/v3/captions?part=snippet&videoId={video_id}&key={api_key}"

# 자막 트랙 정보를 가져옵니다.
response = requests.get(track_info_endpoint)
data = response.json()

# 한국어 자막 트랙 ID를 찾습니다.
korean_track_id = None
for item in data.get("items", []):
    language = item["snippet"]["language"]
    if language == "ko":  # "ko"는 한국어 코드
        korean_track_id = item["id"]
        break

if korean_track_id:
    # 한국어 자막 트랙 ID를 사용하여 자막 내용을 가져옵니다.
    caption_endpoint = f"https://www.googleapis.com/youtube/v3/captions/{korean_track_id}?key={api_key}"

    # 자막 내용을 가져옵니다.
    response = requests.get(caption_endpoint)
    caption_data = response.json()

    # 한국어 자막 스크립트를 출력하거나 필요한 형식으로 가공합니다.
    script_text = caption_data.get("body")
    print(script_text)
else:
    print("Korean captions not available for this video")
```
